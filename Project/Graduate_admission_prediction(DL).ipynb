```python
df = df.drop(["Serial No."], axis=1)
```

```python
df.head()
```

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()   # this scaler is used when we know the min and max vales in features

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.fit_transform(X_test)
```

```python
X_train_scaled
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(7,activation="relu",input_dim=7))

model.add(Dense(7,activation="relu",input_dim=7))

model.add(Dense(1, activation ="linear"))  # activation function will always be linear when working on (regression problems)
```

```python
model.summary()
```

```python
model.compile(loss="mean_squared_error",optimizer="Adam")
```

```python
history = model.fit(X_train_scaled,y_train,epochs=100,validation_split=0.2)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import r2_score
# This is used for Regression problems (Measure of explained variance in target)  ,(-∞ to 1 (higher is better))
r2_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt


plt.plot(history.history["loss"])

plt.plot(history.history["val_loss"])

```
