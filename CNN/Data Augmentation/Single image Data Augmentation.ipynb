```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image 
```

```python
import matplotlib.pyplot as plt
```

```python
img = image.load_img("train/cat.jpg",target_size=(200,200))
```

```python
img
```

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
# img_path = 'train/cat.jpg'  # Replace with your image path
# img = Image.open(img_path)



# Convert the image to an array
img = image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
input_batch.shape
```

```python
i=0
#  flow is for single image , flow to ditectory is for whole dataset 
for output in datagen.flow(input_batch,batch_size=1,save_to_dir = "aug"):
    
    i=i+1
    
    if i==10:
        break
```
